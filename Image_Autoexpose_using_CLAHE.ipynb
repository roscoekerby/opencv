```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = '/content/underexposed image.png'
img = cv2.imread(image_path)

# Convert the image to the LAB color space
lab_img = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

# Apply CLAHE to the L channel
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
lab_img[:, :, 0] = clahe.apply(lab_img[:, :, 0])

# Convert the LAB image back to the BGR color space
output_img = cv2.cvtColor(lab_img, cv2.COLOR_LAB2BGR)

# Save the output image
output_image_path = '/content/exposed image.png'
cv2.imwrite(output_image_path, output_img)

# Convert BGR images to RGB for displaying with plt
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
output_img_rgb = cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB)

# Plot the images side by side
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img_rgb)
plt.axis('off')

# Exposure corrected image
plt.subplot(1, 2, 2)
plt.title('Exposure Corrected Image')
plt.imshow(output_img_rgb)
plt.axis('off')

plt.show()

print(f'Exposure corrected image saved to: {output_image_path}')
```
